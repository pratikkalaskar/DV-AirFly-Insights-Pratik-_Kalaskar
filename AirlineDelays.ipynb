```python
df = df_main.copy()
df
```

```python
df.isnull().sum()
```

```python
# Here 5 records of tail number have null value so i remove it.
df.dropna(subset=['TailNum'], inplace=True)
df=df.reset_index(drop=False)
df
```

```python
# Replace NaN with 0
df['CarrierDelay'] = df['CarrierDelay'].fillna(value=0.0)
df['WeatherDelay'] = df['WeatherDelay'].fillna(value=0.0)
df['NASDelay'] = df['NASDelay'].fillna(value=0.0)
df['SecurityDelay'] = df['SecurityDelay'].fillna(value=0.0)
df['LateAircraftDelay'] = df['LateAircraftDelay'].fillna(value=0.0)

df.head()
```

```python
# Checking Duplicate data
duplicate_data = df[df.duplicated()]
duplicate_data
```

```python
# Converting few columns in proper date format 
df = df.rename(columns={'DayofMonth': 'day', 'Year': 'year', 'Month': 'month'})
df['Date'] = pd.to_datetime(df[['year', 'month', 'day']])
df['DayName'] = df['Date'].dt.day_name()
df.head()
```

```python
# Delete unwanted columns
df = df.drop(['Unnamed: 0','year','month','day','DayOfWeek'], axis=1)
df.head()
```

```python
# Arranging columns properly
df = df.drop(columns=['level_0', 'index'], errors='ignore')
cols = ['Date'] + [col for col in df.columns if col != 'Date']
df = df[cols]

df.head()
```

```python
cols = list(df.columns)
cols.remove("DayName")
cols.insert(1, "DayName")
df_col = df[cols]
df_col.head()
```

```python
df = df_col
df.info()
```

```python
# Reduce int64, float64 and object into int32, float32 and category for better optimization
def optimize_dataframe(df):
    for col in df.columns:
        col_type = df[col].dtypes
        
        if col_type == "int64":
            df[col] = pd.to_numeric(df[col], downcast="integer")
        elif col_type == "float64":
            df[col] = pd.to_numeric(df[col], downcast="float")
    return df
df = optimize_dataframe(df)

cat_cols = ["UniqueCarrier", "Origin", "Dest", "CancellationCode", "TailNum"]
for col in cat_cols:
    df[col] = df[col].astype("category")

df.info()
```

```python
df.head()
```

```python
df_new = df.copy()
df_new
```

```python
df_new.info()
```

```python
df_new.head()
```

```python
df_time = df_new
df_time.head()
```

```python
# Select time-related columns 
time_cols = [c for c in df_time.columns 
             if "Time" in c and str(df_time[c].dtype).startswith(("int", "float"))]

print("Time columns detected:", time_cols)

# Convert each column into proper time format
for col in time_cols:
    df_time[col] = (
        pd.to_datetime(
            df_time[col]
            .fillna(0)                 
            .astype(int)               
            .astype(str)               
            .str.zfill(4)              
            .str.replace("2400","0000"), 
            format="%H%M",
            errors="coerce"
        ).dt.time
    )

df_time.head()
```

```python
df_time.shape
df_time.info()
```

```python
# here delay columns converted into DelayReason and DelayTime column
delay_cols = ["CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]

df_long = df_time.melt(
    id_vars=["Date", "DayName", "DepTime", "CRSDepTime", "ArrTime", "CRSArrTime",
             "UniqueCarrier", "FlightNum", "TailNum", "ActualElapsedTime", "CRSElapsedTime",
             "AirTime", "ArrDelay", "DepDelay", "Origin", "Dest", "Distance",
             "TaxiIn", "TaxiOut", "Cancelled", "CancellationCode", "Diverted"],
    value_vars=delay_cols,
    var_name="DelayReason",
    value_name="DelayTime"
)

# (optional) remove rows with no delay
df_long = df_long[df_long["DelayTime"] > 0]

df_long.head()
```

```python
import calendar

df_new['Date'] = pd.to_datetime(df['Date'])
df_new['Month'] = df_new['Date'].dt.month

monthly_flights = df_new['Month'].value_counts().sort_index()

# Map month numbers to 3-letter names
month_names = [calendar.month_abbr[m] for m in monthly_flights.index]

plt.figure(figsize=(10,6))
plt.plot(month_names, monthly_flights.values, marker='o')
plt.title("Flights Distribution by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()
```

```python
cancelled_flights = df_new['Cancelled'].sum()
print("Cancelled flights:", cancelled_flights)
diverted_flights = df['Diverted'].sum()
print("Diverted flights:", diverted_flights)
```

```python
airline_summary = df_new.groupby("UniqueCarrier").agg(
    TotalFlights = ("Cancelled", "count"),
    Cancelled = ("Cancelled", "sum"),
    Diverted = ("Diverted", "sum")
)

airline_summary["CancelRate_%"] = (airline_summary["Cancelled"] / airline_summary["TotalFlights"]) * 100
airline_summary["DivertRate_%"] = (airline_summary["Diverted"] / airline_summary["TotalFlights"]) * 100

print(airline_summary.sort_values("Cancelled", ascending=False))
airline_summary[["Cancelled","Diverted"]].plot(
    kind="bar", figsize=(12,6), stacked=False
)
plt.title("Cancelled and Diverted Flights by Airline")
plt.ylabel("Number of Flights")
plt.xlabel("Airline")
plt.show()
```

```python
delay_columns = ["CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]

delay_summary = df_new[delay_columns].sum().sort_values(ascending=False)
print(delay_summary)
```

```python
plt.figure(figsize=(8,5))
delay_summary.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Total Delay Minutes by Reason")
plt.ylabel("Total Delay (minutes)")
plt.xlabel("Delay Reason")
plt.show()
```

```python
airline_delay = df_new.groupby("UniqueCarrier")[delay_columns].sum()
print(airline_delay.head())

# Plot stacked bar chart
airline_delay.plot(
    kind="bar", 
    stacked=True, 
    figsize=(12,6),
    colormap="tab20c"
)
plt.title("Delay Reasons by Airline")
plt.ylabel("Total Delay (minutes)")
plt.xlabel("Airline")
plt.legend(title="Delay Type")
plt.show()
```
